```python
def quiz_game():
    questions = [
        {"question": "What is the capital of France?", "answer": "paris"},
        {"question": "What is 2 + 2?", "answer": "4"},
        {"question": "Who wrote 'Hamlet'?", "answer": "shakespeare"},
        {"question": "What is the square root of 16?", "answer": "4"},
        {"question": "What is the largest ocean on Earth?", "answer": "pacific"}
    ]
    score = 0
    for q in questions:
        user_answer = input(q["question"] + " ").strip().lower()
        if user_answer == q["answer"]:
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {q['answer']}")
    print(f"Final Score: {score}/5")
```

```python
def number_swapping():
    try:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print(f"Before swapping: a = {a}, b = {b}")
        a, b = b, a
        print(f"After swapping: a = {a}, b = {b}")
    except ValueError:
        print("Invalid input! Please enter numeric values.")
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def is_armstrong(n):
    num_str = str(n)
    power = len(num_str)
    return sum(int(digit) ** power for digit in num_str) == n

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def number_identifier():
    try:
        num = int(input("Enter a number: "))
        if is_prime(num):
            print(f"{num} is a Prime number.")
        if is_armstrong(num):
            print(f"{num} is an Armstrong number.")
        if is_palindrome(num):
            print(f"{num} is a Palindrome number.")
        if not (is_prime(num) or is_armstrong(num) or is_palindrome(num)):
            print(f"{num} does not meet any of the conditions.")
    except ValueError:
        print("Invalid input! Please enter an integer.")
```

```python
def fibonacci_sequence():
    try:
        n = int(input("Enter the number of Fibonacci terms: "))
        if n <= 0:
            print("Please enter a positive integer.")
            return
        a, b = 0, 1
        for _ in range(n):
            print(a, end=" ")
            a, b = b, a + b
        print()
    except ValueError:
        print("Invalid input! Please enter an integer.")
```

```python
def main():
    while True:
        print("\nChoose an option:")
        print("1. Quiz Game")
        print("2. Number Swapping")
        print("3. Prime/Armstrong/Palindrome Identifier")
        print("4. Fibonacci Sequence Generator")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            quiz_game()
        elif choice == "2":
            number_swapping()
        elif choice == "3":
            number_identifier()
        elif choice == "4":
            fibonacci_sequence()
        elif choice == "5":
            print("Exiting program.")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
